```python
inventory={
            'invid':[1,2,3,4,5],
            'pid':[101,103,105,106,110],
             'color':['blue','brown','chocolate','grey','brown'],
            'price':[10,5,10,40,100]
           }

df_inv=pd.DataFrame(inventory)
df_inv
```

```python
df_inner_merge=pd.merge(df_product,df_inv,how="inner",on=['pid']) # by default it is always inner, it will take intersection of given column
df_inner_merge
```

```python
df_outer_merge=pd.merge(df_product,df_inv,how="outer",on=['pid']) # outer merge will take union of all the rows of dataframe
df_outer_merge
```

```python
df_left_merge=pd.merge(df_product,df_inv,how="left",on=['pid']) # outer merge will take union of all the rows of dataframe
df_left_merge

```

```python
df_right_merge=pd.merge(df_product,df_inv,how="right",on=['pid']) # outer merge will take union of all the rows of dataframe
df_right_merge
```

```python
df_cross_merge=pd.merge(df_product,df_inv,how="cross")
df_cross_merge.head()
```

```python
filepath='/content/Movie_Id_Titles'
df_movie=pd.read_csv(filepath)
df_movie.head()
```

```python
userdatafile='/content/u.data'
df_user=pd.read_csv(userdatafile,sep='\t',names=['userid','item_id','rating','timestamp'])
df_user.head()
```

```python
df_movie_data=pd.merge(df_movie,df_user,on='item_id')
df_movie_data
```

```python
#sort_index()
df_movie_data.sort_values('rating') # sort rating in ascending value
```

```python
df_movie_data.sort_values('rating',ascending=False) # sort rating in descending order
```

```python
df_movie_data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
df_movie_data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
df_avg_rating=pd.DataFrame(df_movie_data.groupby('title')['rating'].mean())
df_avg_rating.head()
```

```python
df_avg_rating['num of reviews']=pd.DataFrame(df_movie_data.groupby('title')['rating'].count())
df_avg_rating.head()
```

```python
moviedata =df_movie_data.pivot_table(index='userid',columns='title',values='rating')
moviedata.head()
```

```python
angry_men_ratings=moviedata['12 Angry Men (1957)']
angry_men_ratings.head(20)
```

```python
similar_movies_to_angrymen =moviedata.corrwith(angry_men_ratings)
similar_movies_to_angrymen
```

```python
df_corr_angrymen=pd.DataFrame(similar_movies_to_angrymen,columns=['correlation'])
df_corr_angrymen.dropna(inplace=True)
df_corr_angrymen
```

```python
df_corr_angrymen.sort_values('correlation',ascending=False).head(10)
```

```python
df_corr_angryment1=df_corr_angrymen.join(df_avg_rating['num of reviews'])
df_corr_angryment1.head()
```

```python
df_corr_angryment1[df_corr_angryment1['num of reviews']>=75].sort_values('correlation',ascending=False).head(10)
```
