```python
import pandas as pd
import numpy as np
```

```python
pdetails={'pid':[101,102,103,104,105,106],
          'pname':['dairy milk','chocobar','5star',np.NaN,np.NaN,'temptation'],
          'quantity':[100,np.NaN,20,50,np.NaN,40]
          }

df_product=pd.DataFrame(pdetails)
df_product
```

```python
df_filled_value=df_product.fillna('No info') # Fill NA/NaN values using the specified method.
df_filled_value
```

```python
df_filled_column_specific=df_product.fillna({'pname':'Out of stock','quantity':df_product['quantity'].mean()})
df_filled_column_specific
```

```python
#method = ffill , bfill
df_forwardfill = df_product.fillna(method='ffill')
df_forwardfill
```

```python
df_backwardfill = df_product.fillna(method='bfill')
df_backwardfill
```

```python
pdetails={'pid':[101,102,103,104,105,106],
          'pname':['dairy milk','chocobar','5star',np.NaN,np.NaN,'temptation'],
          'quantity':[100,np.NaN,20,50,np.NaN,40]
          }

df_product=pd.DataFrame(pdetails)
df_product
```

```python
inventory={
            'invid':[1,2,3,4,5],
            'pid':[101,103,105,106,110],
             'color':['blue','brown','chocolate','grey','brown'],
            'price':[10,5,10,40,100]
           }

df_inv=pd.DataFrame(inventory)
df_inv
```

```python
df_inner_merge=pd.merge(df_product,df_inv,how="inner",on=['pid']) # by default it is always inner, it will take intersection of given column
df_inner_merge
```

```python
df_outer_merge=pd.merge(df_product,df_inv,how="outer",on=['pid']) # outer merge will take union of all the rows of dataframe
df_outer_merge
```

```python
df_left_merge=pd.merge(df_product,df_inv,how="left",on=['pid']) # outer merge will take union of all the rows of dataframe
df_left_merge

```

```python
df_right_merge=pd.merge(df_product,df_inv,how="right",on=['pid']) # outer merge will take union of all the rows of dataframe
df_right_merge
```

```python
df_cross_merge=pd.merge(df_product,df_inv,how="cross")
df_cross_merge.head()
```
